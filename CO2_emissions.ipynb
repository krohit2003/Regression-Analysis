```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score

# Importing the dataset
datas = pd.read_csv('india-emission-1900.csv')
print(datas)
X = datas['Year']
y = datas['Emission']


mymodel = np.poly1d(np.polyfit(X, y, 1))

#myline = np.linspace(1, 22, 100)

plt.scatter(X, y)
plt.plot(X, mymodel(X),color="red")
plt.xlabel("Year")
plt.ylabel("CO2 emissions in tonnes")
plt.title("India CO2 emissions")

plt.show()

print(mymodel)
print('R2 value is =')
print(r2_score(y, mymodel(X)))
#print(y)

```

```python

mymodel = np.poly1d(np.polyfit(X, np.log(y), 1))

#myline = np.linspace(1, 22, 100)

plt.scatter(X, y)
plt.plot(X, np.exp(mymodel(X)),color="red")
plt.xlabel("Year")
plt.ylabel("CO2 emissions in tonnes")
plt.title("India CO2 emissions")

plt.show()
print('R2 score = ')
print(r2_score(np.log(y), mymodel(X)))
#print(y)
```

```python
x_axis=0*y
plt.scatter(y,y- np.exp(mymodel(X)))
plt.plot(y,x_axis,color="red")
plt.xlabel("Actual Value")
plt.ylabel("Error")
plt.title("India CO2 emissions")
```
